```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA



#Part A
#first was open the file, and get each column that is in numbers
#then we take the length of all rows and columns
#we then drop all values that are empty to clean the data
#we also take the z-scores of the numbers, and any data points that are above 3 standard deviations are dropped
#this helps reduce the number of extreme outliers
df = pd.read_csv("CO2_Emissions_Sample.csv")
number_columns = ["CO2_Emissions", "Methane_Emissions", "Nitrous_Emissions", "GDP", "Population"]
dfLength = len(df)
df = df.dropna(subset=number_columns)

z_scores = np.abs(stats.zscore(df[number_columns]))
mask = (z_scores < 3).all(axis=1)
df = df[mask].reset_index(drop=True)

#here we are creating subsets of the data to help visualizaiton on the plotting easier
#so df is all, 2020 is just that year, dense is about half the years, and sparse about a quarter
df2020 = df[df["Year"] == 2020].copy()
dfyearsdense = df[df["Year"].isin([2001, 2003, 2005, 2007, 2010, 2012, 2015, 2017, 2020, 2022, 2024])].copy()
dfyearssparse = df[df["Year"].isin([2001, 2005, 2010, 2015, 2020, 2024])].copy()


#print how many empty lines were deleted
n_removed = dfLength - len(df)
print(f"Number of lines removed during cleaning: {n_removed}")

#first bar chart
#group by top 10 countries in ascending order
#add labels and bar labels
#also add the xlim to allow for space of the figure on Italy
topCountriesCO2 = (df2020.groupby("Country")["CO2_Emissions"]
                .sum()
                .nlargest(10)
                .sort_values(ascending=True))
fig, ax = plt.subplots(figsize=(10,6))
bars = ax.barh(topCountriesCO2.index, topCountriesCO2.values, color="#1a3dba", edgecolor = "white")
ax.set_title("Top 10 Countries by CO2 Emissions in 2020", fontsize = 14, fontweight="bold", pad =12)
ax.set_xlabel("CO2 Emissions (Mt)", fontsize =12)
ax.bar_label(bars, fmt="{:,.0f}", padding=4, fontsize=10)
ax.set_xlim(0, topCountriesCO2.max() * 1.15)
plt.tight_layout()
```

```python

#Second Graph - stacked bar graph
#we index the top countries into a list to make it easier to plot
#then we choose the size and colours for the stack as well as labels for the legend
#then use a for loop to plot the data
topCountriesName = topCountriesCO2.index.tolist()
stack= (df2020[df2020["Country"].isin(topCountriesName)]
           .groupby("Country")[["CO2_Emissions", "Methane_Emissions", "Nitrous_Emissions"]]
           .sum()
           .loc[topCountriesName])
fig, ax = plt.subplots(figsize=(11,6))
colours_stack = ["#4287f5", "#9042f5", "#ec42f5"]
labels_stack = ["CO2", "Methane", "Nitrous"]
left = np.zeros(len(stack))
for col, color, label in zip(stack.columns, colours_stack, labels_stack):
    ax.barh(stack.index, stack[col], left = left, color = color, label = label, edgecolor = "white")
    left += stack[col].values
ax.set_title("Top 10 Countries by CO2, Methane, and Nitrous Emissions in 2020", fontsize = 14, fontweight="bold", pad =12)
ax.set_xlabel("Total Emissions (Mt)", fontsize =12)
ax.legend(loc="lower right", fontsize=10)
plt.tight_layout()
```

```python
#part B
#here we have the scatter plot of GDP vs CO2 emissions, first we take the continents and colour them
#then we plot the data using a for loop and label
#this data I used the df years dense, because with all of the data it seems like too many dots 
#and was difficult to visualize the trends, sparse or just 2020 had much less dots but with so little -
#you could not see trends well either, so I opted for about half the data which I believe is alright
#though I wouldnt say its optimal
fig, ax = plt.subplots(figsize=(10,6))
#continents, australia changed to oceania for the data, and no antarctica as the data has none
continents = ["Asia", "Africa", "North America", "South America", "Europe", "Oceania"]
cont_colours = {
    "Asia": "#942dc4",
    "Africa": "#bd5735",
    "Europe": "#307eab",
    "North America": "#23c21d",
    "Oceania": "#3bd9c4",
    "South America": "#c4bc23",
}

for cont in continents:
    sub = dfyearsdense[dfyearsdense["Continent"] == cont]
    ax.scatter(sub["GDP"], sub["CO2_Emissions"], color = cont_colours[cont], label = cont, alpha=0.8, edgecolors = "white", s=65)
ax.legend(title="Continent", loc="upper right", title_fontsize = 10, fontsize=9)
ax.set_title("GDP vs CO2 Emissions", fontsize = 14, fontweight="bold", pad =12)
ax.set_xlabel("GDP (Billion)", fontsize =12)
ax.set_ylabel("CO2 Emissions (Mt)", fontsize = 12)
plt.tight_layout()
```

```python
#using years sparse, because there were too many bubbles, and it was difficult to see the data
#when using dense there were far too many bubbles, but with sparse I believe it to be a good amount
#you can see some trends, such as green mostly being low or oceania having higher than I expected amount of emissions
#for the code, we first normalize the population by taking each countries population and dividing it by the highest
#this means we can get values between 0 and 1 for population which makes for not wildy small or large bubbles
#the +30 is to allow for even the smallest bubble to have a size of 30
#then we use the for loop to plot the data and format the graph

pop_norm = (dfyearssparse["Population"] / dfyearssparse["Population"].max()) * 2000 + 30

fig, ax = plt.subplots(figsize=(11,7))
for cont in continents:
    mask = dfyearssparse["Continent"] == cont
    ax.scatter(dfyearssparse.loc[mask, "GDP"], dfyearssparse.loc[mask, "CO2_Emissions"], s=pop_norm[mask], 
               color = cont_colours[cont], label = cont, alpha=0.8, edgecolors = "white", linewidths = 0.8)

ax.legend(title="Continent / Population", loc="upper right", title_fontsize = 10, fontsize=9, labelspacing = 1.3)
ax.set_title("GDP vs CO2 Emissions", fontsize = 14, fontweight="bold", pad =12)
ax.set_xlabel("GDP (Billion)", fontsize =12)
ax.set_ylabel("CO2 Emissions (Mt)", fontsize = 12)
plt.ylim(-1000, 38500) #set the ylim to allow for better visualization of the data, slightly negative to see most of the bubbles
                        #and 38500 to see the top bubbles without having outliers make reading ability worse
fig.text(0.5, -0.01, "Larger bubble size = larger population, smaller bubbles = smaller population", ha = "center", fontsize = 9, color="black")

```

```python
#create the correlogram columns, labels, and the matrix using the data
#I used sparse because it seemed to show off slightly more distinctly the correlations
#set the ranges for the graph and plot the matrix
correlogram_columns = ["CO2_Emissions", "GDP", "Methane_Emissions", "Nitrous_Emissions"]
correlogram_labels = ["CO2", "GDP", "Methane", "Nitrous"]
correlogram_matrix = dfyearssparse[correlogram_columns].corr()
fig, ax = plt.subplots(figsize=(7,6))
n = len(correlogram_columns)
cmap = plt.cm.RdBu_r
im = ax.imshow(correlogram_matrix.values, cmap = cmap, vmin=-1, vmax = 1, aspect="auto")

for i in range(n):
    for j in range(n):
        val = correlogram_matrix.values[i, j]
        txt_col = "white" if abs(val) > 0.6 else "black" #determine text colour depending on colour of the squar
        ax.text(j, i, f"{val:.2f}", ha="center", va="center", fontsize = 12, color=txt_col, fontweight = "bold")


ax.set_xticks(range(n)); ax.set_xticklabels(correlogram_labels, fontsize=11)
ax.set_yticks(range(n)); ax.set_yticklabels(correlogram_labels, fontsize=11)
ax.set_title("Correlogram of Emissions vs GDP 2001-2024", fontsize = 14, fontweight="bold", pad =12)
plt.colorbar(im, ax=ax, fraction=0.05, pad = 0.03)
fig.text(0.5, -0.02, "Darker Red = strong positive correlation, Darker blue = strong negative correlation", ha = "center", fontsize = 9, color="black")
```

```python
#continent emissions for 2020, then we do the relative per contintent for emissions to standardize for 100
#plot using for loop and ensure legend isnt covering the right side to allow for good visualization

cont_emissions = (df2020.groupby("Continent")[["CO2_Emissions", "Methane_Emissions", "Nitrous_Emissions"]].sum())
per_cont = cont_emissions.div(cont_emissions.sum(axis=1), axis=0) *100

fig, ax = plt.subplots(figsize=(11,6))
colours_stack = ["#4287f5", "#9042f5", "#ec42f5"]
labels_stack = ["CO2", "Methane", "Nitrous"]
left = np.zeros(len(per_cont))
for col, color, label in zip(per_cont.columns, colours_stack, labels_stack):
    ax.barh(per_cont.index, per_cont[col], left = left, color = color, label = label, edgecolor = "white")
    left += per_cont[col].values
ax.set_title("Top 10 Countries by CO2, Methane, and Nitrous Emissions in 2020", fontsize = 14, fontweight="bold", pad =12)
ax.set_xlabel("Total Emissions (Mt)", fontsize =12)
ax.legend(loc="lower center", fontsize=10)
plt.tight_layout()
```

```python
#pie chart seems easier
#get the total emissions per continent as a sum then plot on the pie chart
#this is where we see the possible errors in the data clearly
#maybe north America is really that low or my cleaning of the data was not nearly sufficient
#or the data is not accurate
total_emissions_cont = (df2020.groupby("Continent")[["CO2_Emissions", "Methane_Emissions", "Nitrous_Emissions"]]
                        .sum()
                        .sum(axis=1))
fig, ax = plt.subplots(figsize=(8, 7))
wedges, texts, autotexts = ax.pie(total_emissions_cont, labels = total_emissions_cont.index, autopct="%1.1f%%",
                                  colors = [cont_colours[c] for c in total_emissions_cont.index], 
                                  wedgeprops={"edgecolor": "white"})
ax.set_title("Total Emissions by Continent in 2020", fontsize = 14, fontweight="bold", pad =12)
```

```python

#set the 5 countries we are doing the chart on, I just chose relatively randomly trying to get a good amount of the world
#set country colours, then we take the mean of co2 emissions and plot
#plot with the dotted lines for the raw data, and the solid lines for the smoothed averages
#set labels
five_countries = ["Russia", "India", "Canada", "France", "Japan"]
country_colours = {
    "Russia": "#942dc4",
    "India": "#bd5735",
    "Canada": "#307eab",
    "France": "#23c21d",
    "Japan": "#3bd9c4"
}
ts_data = (df[df["Country"].isin(five_countries)]
           .groupby(["Country", "Year"])["CO2_Emissions"]
           .mean()
           .reset_index())
fig, ax = plt.subplots(figsize = (10,6))
for country, grp in ts_data.groupby("Country"):
    grp_sorted = grp.sort_values("Year")
    color = country_colours[country]
    ax.plot(grp_sorted["Year"], grp_sorted["CO2_Emissions"], color = color, alpha = 0.5, linewidth = 1.4, linestyle = "--")
    ma = grp_sorted["CO2_Emissions"].rolling(window=5, min_periods=1).mean()
    ax.plot(grp_sorted["Year"], ma,
            color=color, linewidth=2.5, label=country)
ax.set_title("CO2 Emissions from 2000-2024, for 5 Countries", fontsize = 14, fontweight="bold", pad =12)
ax.set_xlabel("Year", fontsize =12)
ax.set_ylabel("Total Emissions (Mt)", fontsize =12)
ax.set_xticks(range(2000, 2025, 2))
fig.text(0.5, -0.02, "Dashes are raw data and solid lines are 5 year averages.", ha = "center", fontsize = 9, color="black")
ax.legend(title="Country", fontsize=10, title_fontsize=11)

```

```python
#1: filter the data to 2020, then take the mean of all the columns and print
# we now have the continents represented by the feature vectors of 5 values
pca_features = ["CO2_Emissions", "Methane_Emissions", "Nitrous_Emissions", "GDP", "Population"]
pca_df = (df[df["Year"] == 2020]
          .dropna(subset=pca_features)
          .groupby("Continent")[pca_features]
          .mean())
print(pca_df.round(2).to_string())

#set up the sca feature extraction
#print the results
scaler = StandardScaler()
x_scaled = scaler.fit_transform(pca_df)

pca = PCA(n_components = 2)
x_pca = pca.fit_transform(x_scaled)
var_exp = pca.explained_variance_ratio_ * 100
print(f"\nExplained Variance --> PC1: {var_exp[0]:.1f}%  |  PC2: {var_exp[1]:.1f}%")
print(f"Total explained: {sum(var_exp):.1f}%")


#plot the results, large dots, add the labels to the points
fig, ax = plt.subplots(figsize=(9,7))
for i, continent in enumerate(pca_df.index):
    color = cont_colours.get(continent)
    ax.scatter(x_pca[i, 0], x_pca[i, 1],
               color=color, s=220, edgecolors="white", linewidths=1.5,
            label=continent)
    ax.annotate(continent,
                xy=(x_pca[i, 0], x_pca[i, 1]),
                xytext=(8, 6), textcoords="offset points",
                fontsize=10, fontweight="bold", color=color)

#add the gridlines at 0,0 and labels
ax.axhline(0, color="grey", linewidth=0.8, linestyle="--", alpha=0.6)
ax.axvline(0, color="grey", linewidth=0.8, linestyle="--", alpha=0.6)
ax.set_xlabel(f"PC1 ({var_exp[0]:.1f}% variance explained)", fontsize=12)
ax.set_ylabel(f"PC2 ({var_exp[1]:.1f}% variance explained)", fontsize=12)
```
