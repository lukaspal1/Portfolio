```python
#first we have to get our imports and set stopwords
#then we will set some defaults as well as test to make sure data sheet is working
#get the columns and missing values as well as duplicate rows

import micropip
await micropip.install(['pandas', 'matplotlib', 'seaborn', 'wordcloud', 'plotly', 
                        'stop_words', 'scipy', 'nbformat'])

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
import re, string
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud
from collections import Counter
from scipy import stats
from stop_words import get_stop_words

#set a default font size
plt.rcParams['font.size'] = 11 

print("\n---------------------------------------------------\n")
df = pd.read_csv("Climate_Tweets_Sample.csv")
print('Shape:', df.shape)
print('\nColumn dtypes:')
print(df.dtypes)
print('\nMissing values per column:')
print(df.isnull().sum())
print('\nDuplicate rows:', df.duplicated().sum())
df.head()
```

```python
#Part A
#cleaning and sentiment overview
#first get the proper date parsing
#take the dates, and convert into a proper format using pandas to infer the current format
df['date'] = pd.to_datetime(df['date'], errors='coerce')

print(f'Date range : {df["date"].min().date()} - {df["date"].max().date()}')
print(f'Unparseable dates: {df["date"].isna().sum()}')

print("\n---------------------------------------------------\n")
#then we drop any duplicates
#print pre and post drop
pre_n = len(df)
df = df.drop_duplicates()
print(f'Rows before dropping duplicates : {pre_n}')
print(f'Rows after dropping duplicates : {len(df)}')

print("\n---------------------------------------------------\n")

#finally we will handle the missing values
#missing values are only in like and retweet count as we found earlier
#we fill with 0s, as missing data likely means that there were no likes or no retweets
#if not, its plausable that there are 0 likes or retweets as well
#we could have used drop but instead I think it would be good to assume some posts just have 
#no likes or retweets
print(f'Missing values before: \n{df[["like_count", "retweet_count"]].isnull().sum()}')

df['like_count'] = df['like_count'].fillna(0).astype(int)
df['retweet_count'] = df['retweet_count'].fillna(0).astype(int)

print(f'\nMissing values After: \n{df[["like_count", "retweet_count"]].isnull().sum()}')
```

```python
#Part A - Sentiment Visualization
#create a bar chart showing overall distribution of tweet sentiments
#first we set colours for the sentiments
#then we will get the counts for the sentiments and re-index them to prositive neutral and negative
COLOURS = {'positive': 'green', 'neutral': 'blue', 'negative': 'red'}
sentiment_counts = df['sentiment'].value_counts().reindex(['positive', 'neutral', 'negative'])

#plot the data using the colours, 
fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(sentiment_counts.index, sentiment_counts.values,
              color=[COLOURS[s] for s in sentiment_counts.index])

#put labels on each bar showing the number of tweets as well as the percentage
#of tweets for that sentiment
for bar, val in zip(bars, sentiment_counts.values):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() +10,
            f'{val}\n({val / len(df) * 100:.1f}%)',
            ha='center', va='bottom', fontsize=10)

#labels, ylim to 1.2 to give space for counts
ax.set_title('Overall Distribution of Tweet Sentiments', fontweight='bold', pad=10)
ax.set_xlabel('Sentiment')
ax.set_ylabel('Number of Tweets')
ax.set_ylim(0, sentiment_counts.max() * 1.2)
plt.show()

#overall sentiment seems to be negative, otherwise is almost even.
#common throughout media to have negative sentiments because they get more clicks
```

```python
#Sentiment Visualization - top 10 countries

#first we get the top 10 countries by amount of tweets, then create a list
#the list is sorted by highest to lowest 10 countries using .head(10)
#then we get take all the tweets sent by those top 10 countries
top10_countries = df['country'].value_counts().head(10).index.tolist()
df_top10 = df[df['country'].isin(top10_countries)]

#we take the countries and tweets and group them by the sentiment and country
top10_set = (df_top10
         .groupby(['country', 'sentiment'])
         .size()
         .unstack(fill_value=0)
         [['positive', 'neutral', 'negative']])

#then reorder the countries to show to the x axis sorted by tweet volume instead of alphabetically
top10_set = top10_set.loc[df_top10['country'].value_counts().index]
#plot the data using a bar graph, same colours, and increase the bar width
fig, ax = plt.subplots(figsize=(12, 5))
top10_set.plot(kind='bar', ax=ax,
           color=[COLOURS[s] for s in top10_set.columns],
            width=0.75)

#set labels, and for tick labels rotate 30 degrees for readabilty
ax.set_title('Sentiment Distribution – Top 10 Countries by Tweet Volume',
             fontweight='bold', pad=10)
ax.set_xlabel('Country')
ax.set_ylabel('Number of Tweets')
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
ax.legend(title='Sentiment', loc='upper right')
plt.show()

#we can see that most countries are similar in volume of tweets, and sentiment is generally similar
#notable bars we can see are Mexico and Saudi Arabia with relatively high negative sentiment
#compared to the rest of the countries on the graph
```

```python
#Part B - text preprocessing
stop_words = set(get_stop_words('en'))

#adding extra stopwords that are relevant to the data
stop_words.update([
    'rt', 'amp', 'via', 'would', 'could', 'also', 'get', 'new', 'one',
    'us', 'check', 'figures', 'details', 'attached', 'incoming',
    'says', 'said', 'fyi', 'breaking', 'reminder', 'opinion', 'update',
    'continues', 'discussion', 'today', 'heads', 'learn', 'share',
    'stay', 'informed', 'source', 'comments', 'report', 'overview',
    'policies', 'raise', 'raised', 'boost', 'read', 'full', 'latest', 
    'data', 'responsibly', 'grow', 'concerns', 'experts',
    'debate', 'signs', 'trends', 'communities', 'encouraging', 'alarm',
    'impacts'
])
#testing to make sure stop words works
text = "The quick brown fox jumps over the lazy dog"
words = text.lower().split()
filtered_words = [word for word in words if word not in stop_words]
print(filtered_words)  # ['quick', 'brown', 'fox', 'jumps', 'lazy', 'dog']
print("\n---------------------------------------------------\n")

#create a function to clean the text
#first set to lowercase
#then we remove punctuation such as colons and semi-colons
#then we remove the rest of punctuation like periods or exclamation marks
#then we split the text into a list of individual words
#then we remove stopwords and make sure any words less than 3 characters get filtered incase its not in stopwords
#then join the tokens back into clean text instead of list of words
def clean_text(text: str) -> str:
    text = str(text).lower()
    text = re.sub(r'[;:,]', ' ', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
    return ' '.join(tokens)

#apply the function to the text
#showcase a cleaned tweet
df['clean_text'] = df['text'].apply(clean_text)
print('Text has been cleaned')
print('Original :', df['text'].iloc[1])
print('Cleaned  :', df['clean_text'].iloc[1])

#didnt use lemmatization :)
```

```python
#Keyword and word cloud analysis

#take all the tokens and join them into 1 giant string
#split that string into a list of words
#then count all the words and find the 20 most common with counts
#then unzip them into 2 variables
all_tokens = ' '.join(df['clean_text']).split()
top20 = Counter(all_tokens).most_common(20)
words20, counts20 = zip(*top20)

#plot the data
fig, ax = plt.subplots(figsize=(10, 5))
ax.barh(list(words20)[::-1], list(counts20)[::-1])
#set labels and add the counts on the words
ax.set_title('Top 20 Most Frequent Terms', fontweight='bold', pad=10)
ax.set_xlabel('Frequency')
ax.set_ylabel('Term')
for i, v in enumerate(list(counts20)[::-1]):
    ax.text(v + 1, i, str(v), va='center', fontsize=9)
plt.show()
#There were many stopwords that had to be added to the list, but I am satisfied what words we have now
#we can see the general sentiments of worsening, setbacks, as well as progress and promising
#then we see a few phrases like global warming, carbon tax, and air quality
#overall sentiments are mixed and many are concerned with global warming and renewable energy
#as well as carbon tax among others
```

```python
positive_tweets = ' '.join(df.loc[df['sentiment'] == 'positive', 'clean_text'])
#using a background colour of black it makes it easy to see all the words, 
#when white some where near impossible to read
wc_positive = WordCloud(width=900, height=450, background_color='black',
                   colormap= 'Greens',
                   collocations=False).generate(positive_tweets)

fig, ax = plt.subplots(figsize=(11, 5))
ax.imshow(wc_positive, interpolation='bilinear')
ax.axis('off')
ax.set_title('Word Cloud – Positive Climate Tweets', fontsize=14, fontweight='bold', pad=10)
plt.show()
# we see obviously similar words to the positive ones in the top 20, but we also see
#global warming which is interesting to be in positive, or heatwave, or drought, or wildfire
#it is likely due to some tweets saying that there is a decrease in wildfires or actions on these things
```

```python
negative_tweets = ' '.join(df.loc[df['sentiment'] == 'negative', 'clean_text'])
#similarly, some reds are very light and almost invisible with white background
wc_negative = WordCloud(width=900, height=450, background_color='black',
                   colormap= 'Reds',
                   collocations=False).generate(negative_tweets)
#otherwise copy paste the code and change to negative
fig, ax = plt.subplots(figsize=(11, 5))
ax.imshow(wc_negative, interpolation='bilinear')
ax.axis('off')
ax.set_title('Word Cloud – Negative Climate Tweets', fontsize=14, fontweight='bold', pad=10)
plt.show()
#we see many similar words, such as heatwave or green which is interesting,
#both sides having different sentiments using the same words
#biggest words obviously being worsening, and setbacks
```

```python
#part C - geospacial visualization
#aggregate the countries with summary columns of tweet count and percentage of positive tweets
#print off total countries to get an idea of how many are in the dataset
aggregation = (df.groupby('country').agg(
                    tweet_count = ('tweet_id', 'count'),
                    percent_positive = ('sentiment', lambda x: (x == 'positive').mean() * 100))
                    .reset_index().round({'percent_positive': 1}))
print(f'Total Countries in the Dataset: {len(aggregation)}\n')
#I used reset index drop=true because the index numbers on the left could be confusing
#otherwise quite straighforward, ascending is false to keep it sorted by tweet count instead of pos tweets
aggregation.sort_values('tweet_count', ascending=False).reset_index(drop=True).head(10)

```

```python
#Choropleth Implementation

#getting a warning for country names as location mode
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)

#create choropleth
#use blues because its a good neutral colour
fig_map = px.choropleth(
    aggregation,
    locations              = 'country',
    locationmode           = 'country names',
    color                  = 'tweet_count',
    hover_name             = 'country',
    hover_data             = {'tweet_count': True, 'percent_positive': ':.1f'},
    color_continuous_scale = 'Blues',
    labels                 = {'tweet_count': 'Tweet Count', 'percent_positive': '% Positive'},
    title                  = 'Climate Tweet Volume by Country (2021–2023)'
)
#update the layout
fig_map.update_layout(
    geo = dict(showframe=False, showcoastlines=True, projection_type='natural earth'),
    coloraxis_colorbar = dict(title='# of Tweets', thickness=15),
    margin = dict(l=0, r=0, t=50, b=0),
    height = 480
)
fig_map.show()
#we can see that its relatively random which countries have the most tweets,
#such as New Zealand and Mexico. Also Missing many countries with tweets
```

```python
#Part D - Monthly nagative sentiment rate and 95% ci
#take the date column and covert it into monthly periods
df['month_period'] = df['date'].dt.to_period('M')

#take the month periods and aggregate them with sentiment and count groups
monthly = (df.groupby('month_period').agg(total = ('sentiment', 'count'),
                  n_negative = ('sentiment', lambda x: (x == 'negative').sum()))
             .reset_index()
             .sort_values('month_period'))

#then we get the proportion of tweets that are negative per month
#get the confidence interval margin using the regular formula
#then we get the upper and lower confidence intervals
#finally convert the periods to timestamps and display the first year of timestamps
monthly['rate'] = monthly['n_negative'] / monthly['total']
monthly['margin'] = 1.96 * np.sqrt(monthly['rate'] * (1 - monthly['rate']) / monthly['total'])
monthly['ci_high'] = monthly['rate'] + monthly['margin']
monthly['ci_low'] = monthly['rate'] - monthly['margin']
monthly['date_ts'] = monthly['month_period'].dt.to_timestamp()
monthly.head(12)
```

```python
#plot the data
#set the size of the graph, then fill the inside with the red colour
#label that the 95% CI
#then we plot the exact data points of negative rates
#then we get the mean, and create the dotted line to show the average
#titles, labels, and rotate the dats for easier readability

#
fig, ax = plt.subplots(figsize=(13,5))

ax.fill_between(monthly['date_ts'], monthly['ci_low'], monthly['ci_high'], 
                color = 'red', label='95% CI')
ax.plot(monthly['date_ts'], monthly['rate'],
        color='black', linewidth=2, marker='o', markersize=4,
        label='Monthly Negative Rate')
ax.axhline(monthly['rate'].mean(), color='black', linestyle='--', linewidth=1,
           label=f"Overall mean = {monthly['rate'].mean():.2f}")
ax.set_title('Monthly Negative Sentiment Rate with 95% Confidence Interval',
             fontweight='bold', pad=10)
ax.set_xlabel('Month')
ax.set_ylabel('Proportion Negative')
ax.set_ylim(0, 0.75)
ax.legend(framealpha=0.85, loc='lower right')
fig.autofmt_xdate(rotation=30)
plt.show()
#we see that the overall negative sentiment rate is relatively stable,
#fluctuating between about 15% above or below the overall mean.
#there is a noticable drop in early 2021 and late 2022 likely from large 
#positive events that would decrease the negative sentiment
```

```python
#scatter plot :)
#take the top 5 countries t with the most tweets
#colored by sentiment

#essentially doing the same as the sentiment visualization
top5 = df['country'].value_counts().head(5).index.tolist()
df_scatter = df[df['country'].isin(top5)].copy()

#plot the scatter
fig, ax = plt.subplots(figsize=(9, 6))

for sent, grp in df_scatter.groupby('sentiment'):
    ax.scatter(grp['retweet_count'], grp['like_count'],
               color=COLOURS[sent], alpha=0.6, s=50,
              label=sent.capitalize())
#labels, set x and y lims to cut out some extreme outliers
ax.set_title('Likes Vs Retweets Coloured by Sentiment \n(Countries: New Zealand, Thailand, Mexico, Spain, Denmark)',
             fontweight='bold', pad=10)
ax.set_xlabel('Retweet Count')
ax.set_ylabel('Like Count')
ax.set_xlim(-10, 125)
ax.set_ylim(-10, 375)

ax.legend(title='Sentiment', framealpha=0.85)
plt.show()
#almost no pattern that is discernable from the graph, there are a few outliers that show
#negative sentiment can get larger like and retween counts, however we also see the large positive
#tweet. Otherwise essentially all bubbles are spread relatively evenly throughout the graph
```
